```python
import glob#import .xml file from my files into Jupyter
files = glob.glob("*xml")
print(files)#successfully imported one of Chekhov's short stories for analysis
```

```python
f = open("Смерть чиновника.xml", "r")#open file in read mode
textString = f.read()
f.close()

with open("Смерть чиновника.xml", "r") as f: 
    chekhovString = f.read()
```

```python
start = chekhovString.find("В")#define start and end points of story text
end = chekhovString.find("помер")
chinovnikString = chekhovString[start:end]
```

```python
import nltk
```

```python
chinovnikStringLowercase = nltk.word_tokenize(chinovnikString.lower())#make text lowercase
chinovnikStringTokensLowercase = [word for word in chinovnikStringLowercase if word[0].isalpha()]#tokenize
chinovnikStringTokensLowercaseFreqs = nltk.FreqDist(chinovnikStringTokensLowercase)#frequency dist
chinovnikStringTokensLowercaseFreqs.tabulate(20)
```

```python
chinovnikStringTokensLowercaseFreqs.plot(25, title="Top Frequency Word Tokens in Смерть Чиновника")
```

Looking at this visualization, I see it has done a good job of identifying the most frequenctly used words. As I am using an xml file, I note that the second most frequently used word is actually the "p" tag, so I should try to find a way to filter this out were I to perform further analysis. As the document is an xml file, I could probably use xpath, but since this is the only tag, it might actually be easier just to add "p" to the stop word list. 

```python
chinovnikStringTokensLowercaseLengths = [len(w) for w in chinovnikStringTokensLowercase]#count length of words
print("first five words: ", chinovnikStringTokensLowercase[:5])
print("first five word lengths: ", chinovnikStringTokensLowercaseLengths[:5])
```

```python
nltk.FreqDist(chinovnikStringTokensLowercaseLengths).plot()#This graph shows that shorter words are used more frequently
```

```python
chinovnikStringTokensLowercaseLengthsFreqs = list(sorted(nltk.FreqDist(chinovnikStringTokensLowercaseLengths).items()))
chinovnikStringTokensLowercaseLengthsFreqs#This shows the frequency of words that contain 1 letter, 2 letters, etc. 
```

```python
chinovnikStringTokensLowercaseLengths = [f[0] for f in chinovnikStringTokensLowercaseLengthsFreqs]
chinovnikStringTokensLowercaseLengthsValues = [f[1] for f in chinovnikStringTokensLowercaseLengthsFreqs]
plt.plot(chinovnikStringTokensLowercaseLengths, chinovnikStringTokensLowercaseLengthsValues)
plt.xlabel('Word Length')
plt.ylabel('Word Count')
```

Above we can see Chekhov's Characteristic Curve. Interestingly, it shows a dip in 4 letter words and a jump in five and letter words. I wonder if this is related to the grammatical structure of Russian and if this would be the same for other Russian authors, or if it would hold steady across Chekhov's work. 

```python
stopwords = nltk.corpus.stopwords.words("Russian")#add Russian stop words
chinovnikStringTokensLowercase = [word for word in chinovnikStringTokensLowercase if word not in stopwords]
chinovnikStringTokensLowercaseFreqs = nltk.FreqDist(chinovnikStringTokensLowercase)
chinovnikStringTokensLowercaseFreqs.most_common(20)#show top 20 most frequent terms
```

```python
chinovnikStringTokensLowercaseFreqs.plot(20, title ="Top Frequency Content Terms in Смерть Чиновника")
```
