Okay, so I have confirmed that urllib.request now works and that I can pull up the necessary HTML code. Feeling stressed because I am really behind now, but I am doing my best. I already know how to use BeautifulSoup pretty well, I just need to work through it. 

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uopen
source = uopen('https://www.cbc.ca/')
CBCsoup = bs(source, features = 'lxml')

#print(CBCsoup.prettify())
#I printed the soup using the prettify function to help me look through the HTML for the relevant tags and CSS indicators
```

I have used the Prettify function of BeautifulSoup to look for relevant tags and have identified that headlines seem to be embedded within h3 tags that belong to the class "headline". I will use Beautiful Soup's search function to see if this is correct. 

```python
CBCsoup.find_all("h3")
```

Okay, so this is looking good. I have a clean list of headlines (as well as a few footer headings I don't need). So I just need to drill down one level to the class="headline" level. Here it goes!

```python
CBCsoup.find_all("h3", class_="headline")
```

Great! Okay, the next step is to extract just the text without the HTML tags (and not get distracted by the headlines). So I need to use .get_text() I think. Let's try it. I think I need to put it into a for loop. 

```python
for text in CBCsoup.find_all("h3", class_="headline"):
    print(text.get_text())
```

YES! It worked! Okay. I'd probably clean this up a bit, but it looks really good. So let's move on to FoxNews. Ah yes, there's Trump. So I'm going to use the same method I used before. I need to create a new soup, print it in the Prettify form, and isolate the HTML tags. 

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uopen
source = uopen('https://www.foxnews.com')
Foxsoup = bs(source, features = 'lxml')

#print(Foxsoup.prettify())
```

It looks like the headlines are in the h2 tag with the class "title". So I'm going to see what happens when I get search for h2 class="title". 

```python
Foxsoup.find_all("h2", class_="title")
```

Okay, so this is good, but embedded within the h2 tags are also the links to the individual stories. So I need to delete those and only get the text. Maybe I can just use the same method I used above with .get_text(). 

```python
for text in Foxsoup.find_all("h2", class_="title"):
    print(text.get_text())
```
